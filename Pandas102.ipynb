#ข้อ 1

````python
import pandas as pd
import os

from google.colab import drive
drive.mount('/content/drive')
path = '/content/drive/MyDrive/Dataviz2023'
````

````python
sep_df = pd.read_excel('/content/drive/MyDrive/DataViz23Data/sep65.xlsx', header=8, skipfooter=26)
sep_df = sep_df.iloc[-2:,:15]
sep_df['เดือน-ปี'] = len(sep_df)*['sep-65']
sep_df
````

````python
oct_df = pd.read_excel('/content/drive/MyDrive/DataViz23Data/oct65.xlsx', header=9, skipfooter=27)
oct_df = oct_df.iloc[-2:,:15]
oct_df['เดือน-ปี'] = len(oct_df)*['oct-65']
oct_df
````

````python
sep_oct_df = pd.concat([sep_df, oct_df])
sep_oct_df
````

````python
fix_sep = sep_oct_df[(sep_oct_df['รายการ']=='ประจำ')&( sep_oct_df['เดือน-ปี']=='sep-65')]['งบประมาณ']
invest_sep = sep_oct_df[(sep_oct_df['รายการ']=='ลงทุน')&( sep_oct_df['เดือน-ปี']=='sep-65')]['งบประมาณ']
fix_oct = sep_oct_df[(sep_oct_df['รายการ']=='ประจำ')&( sep_oct_df['เดือน-ปี']=='oct-65')]['งบประมาณ']
invest_oct = sep_oct_df[(sep_oct_df['รายการ']=='ลงทุน')&( sep_oct_df['เดือน-ปี']=='oct-65')]['งบประมาณ']
if int(fix_sep) > int(invest_sep): print(f'ในเดือนกันยายน งบประมาณประจำ สูงกว่างบประมาณลงทุนอยู่',int(fix_sep)-int(invest_sep),'บาท')
else: print (f'ในเดือนกันยายน งบประมาณลงทุน สูงกว่างบประมาณประจำอยู่',int(invest_sep)-int(fix_sep),'บาท')

if int(fix_oct) > int(invest_oct): print(f'ในเดือนตุลาคม งบประมาณประจำ สูงกว่างบประมาณลงทุนอยู่', int(fix_oct)-int(invest_oct),'บาท')
else: print (f'ในเดือนตุลาคม งบประมาณลงทุน สูงกว่างบประมาณประจำอยู่', int(invest_oct)-int(fix_oct),'บาท')
````

#ข้่อ 2


````python
total_sep = int(fix_sep) + int(invest_sep)
total_oct = int(fix_oct) + int(invest_oct)
if total_sep > total_oct : print(f'งบรวมเดือนกันยายนมากกว่าเดือนตุลาคมอยู่',(total_sep - total_oct), 'บาท')
else: print(f'งบรวมเดือนตุลาคมมากกว่าเดือนกันยายนอยู่',(total_oct - total_sep), 'บาท')
````

#ข้อ 3

````python
sep_oct_df = pd.concat([sep_df, oct_df])
sep_oct_df.rename(columns=lambda x: x.replace('\n', ''), inplace=True)
sep_oct_df.rename(columns=lambda x: x.replace('*/-', ''), inplace=True)
sep_oct_df.fillna('-', inplace = True)
sep_oct_df
````

#ข้อ 4

````python
PO_sep = sep_oct_df [( sep_oct_df['เดือน-ปี']=='sep-65')]['การใช้จ่าย (PO+เบิกจ่าย)'].sum()
PO_oct = sep_oct_df[( sep_oct_df['เดือน-ปี']=='oct-65')]['การใช้จ่าย (PO+เบิกจ่าย)'].sum()
if int(PO_sep) > int(PO_oct): print(f'เดือนกันยายน เบิกจ่ายรวมมากกว่า เดือนตุลาคมอยู่',int(PO_sep - PO_oct),'บาท')
else: print(f'เดือนตุลาคม เบิกจ่ายรวมมากกว่า เดือนกันยายนอยู่',int(PO_oct - PO_sep),'บาท')
````
